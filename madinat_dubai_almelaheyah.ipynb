```python
def clean_transactions_data(df):
    # Convert 'instance_date' to datetime
    df['instance_date'] = pd.to_datetime(df['instance_date'], errors='coerce', format='%d-%m-%Y')
    
    # Drop irrelevant columns
    df.drop(columns=['trans_group_ar', 'procedure_name_ar', 'property_type_ar', 'property_sub_type_ar',
                      'property_usage_ar', 'reg_type_ar', 'area_name_ar', 'building_name_ar', 'project_name_ar',
                      'master_project_ar', 'nearest_landmark_ar', 'nearest_metro_ar', 'nearest_mall_ar',
                      'rooms_ar', 'rent_value', 'meter_rent_price'], inplace=True, errors='ignore')
    
    return df


def clean_rent_contracts_data(df):
    # Convert 'contract_start_date' and 'contract_end_date' to datetime
    df['contract_start_date'] = pd.to_datetime(df['contract_start_date'], errors='coerce', format='%d-%m-%Y')
    df['contract_end_date'] = pd.to_datetime(df['contract_end_date'], errors='coerce', format='%d-%m-%Y')
    
    # Drop irrelevant columns
    df.drop(columns=['contract_reg_type_ar', 'ejari_bus_property_type_ar', 'ejari_property_type_ar',
                      'ejari_property_sub_type_ar', 'property_usage_ar', 'project_name_ar', 'master_project_ar',
                      'nearest_landmark_ar', 'nearest_metro_ar', 'nearest_mall_ar', 'tenant_type_ar'],
            inplace=True, errors='ignore')
    
    return df

# Apply cleaning functions
mina_rashid_tx = clean_transactions_data(mina_rashid_tx)
mina_rashid_rent = clean_rent_contracts_data(mina_rashid_rent)
```
